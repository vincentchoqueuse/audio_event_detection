# Enveloppe Detection

import numpy as np
import matplotlib.pyplot as plt
from scipy.io.wavfile import read

```python
def plot_waveform(Fs,data,label="signal"):
    t = np.arange(len(data))/Fs
    plt.plot(t,data,label=label)
    plt.xlabel("time (seconds)");
```

## Load wavefile


```python
filename_list = ["wav/test.wav","wav/test2.wav","wav/test3.wav"]
filename = filename_list[1]
```

```python
Fs, audio_data_full = read(filename)
audio_data = (1/2**15) * audio_data_full[:,0]
plot_waveform(Fs,audio_data,"signal")
```

## Envelope Extraction

To extract the envelope, we can first compute the square root of each sample and then apply a lowpass iir filter (controled by the coefficient $\alpha$). In the following, the algorithm is implemented using a for loop function (as it will be implemented using a low level programming language).

```python
def envelope_detector(audio_data,alpha):

    N = len(audio_data)
    envelope = np.zeros(N)
    previous_value = 0

    for index in range(N) : 
        envelope[index] = (audio_data[index]**2)+alpha*previous_value
        previous_value = envelope[index]
        
    return envelope
```

```python
alpha = 0.9995 #manual tuning
envelope = envelope_detector(audio_data,alpha)
plot_waveform(Fs,envelope,"env")
```

## Event Trigger

To detect an audio event, we simple use a threshold detector based on the envelope.

```python
threshold = 100  #manual tuning
trigger = 1*(envelope>threshold)

plot_waveform(Fs,audio_data,"audio")
plot_waveform(Fs,trigger,"trigger")

plt.legend();
```

```python
plot_waveform(Fs,trigger,"trigger")
plt.xlim([0,1.1])
```
