# Audio with Python

```python
import numpy as np
import matplotlib.pyplot as plt

from scipy.io.wavfile import read
from IPython.display import Audio, display
from ipywidgets import interact
```

## Load wavefile in the wav folder

In the first cell, we load a stereo wavfile with the read function of the `scipy.io.wavfile` module. Then we display the sampling rate and the size of data.

```python
filename_list = ["wav/test.wav","wav/test2.wav","wav/test3.wav"]
filename = filename_list[2]
```

```python
Fs, audio_data_full = read(filename)
N = audio_data_full.shape[0] #total time in samples
N_sec = N/Fs #total time in second

print("sampling frequency = {} samples/second".format(Fs))
print("data length : {}".format(audio_data_full.shape))
```

The wavfile is a stereo file, for simplicity, we only keep the left channel. Audio data are coded as 16bits integers. In the following we normalized the value of the data to the range [-1,1].

```python
norm = 1/2**15
audio_data = norm * audio_data_full[:,0]
```

We can display the audio file using the matplotlib `plot` function

```python
t = np.arange(N)/Fs # create time vector (index of the array divided by the sampling rate)
plt.plot(t,audio_data)
plt.xlabel("time (seconds)");
```

## Jupyter tools for audio data

### Audio visualisation using widget

```python
@interact(t_max=(0.1,N_sec,0.1))
def show_audio(t_max=8):
    N_max = int(t_max *Fs)
    plt.plot(t[:N_max],audio_data[:N_max])
    plt.xlabel("time (seconds)");
```

## Play File with Python

```python
display(Audio(data=audio_data, rate=Fs))
```
